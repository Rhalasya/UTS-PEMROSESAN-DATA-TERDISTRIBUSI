```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset dari file CSV
data = pd.read_csv('housing.csv')
data
```

```python
# pemilihan feature
data_train = data.drop(["ocean_proximity"],axis=1)
data_train
```

```python
jumlah_data = len(data_train)
rataan_data = data_train.mean()
median_data = data_train.median()
mean_data = data_train.mean()
```

```python
print("Jumlah data:", jumlah_data, "\n")
print("Rataan data:\n",rataan_data, "\n")
print("Median data:\n",median_data, "\n")
print("Mean data:\n \n",mean_data)
```

```python
# Data yang akan divisualisasikan
labels = ['Jumlah Data']
values = [jumlah_data]

# Membuat diagram batang
plt.figure(figsize=(8, 6))
plt.bar(labels, values, color=['blue', 'green', 'orange'])
plt.xlabel('Statistik Data')
plt.ylabel('Nilai')
plt.title('Statistik Data Training Set')
plt.show()
```

```python
# Mengubah data menjadi format yang cocok untuk visualisasi
rataan_data = rataan_data.reset_index()
rataan_data.columns = ['Fitur', 'Rata-rata']

# Membuat grafik menggunakan Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='Rata-rata', y='Fitur', data=rataan_data, palette='viridis')
plt.xlabel('Rata-rata')
plt.ylabel('Fitur')
plt.title('Rata-rata Tiap Fitur dalam Dataset Housing')
plt.show()
```

```python
# Mengubah data menjadi format yang cocok untuk visualisasich
median_data = median_data.reset_index()
median_data.columns = ['Fitur', 'Median']

# Membuat grafik menggunakan Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='Median', y='Fitur', data=median_data, palette='coolwarm')
plt.xlabel('Median')
plt.ylabel('Fitur')
plt.title('Median Tiap Fitur dalam Dataset Housing')
plt.show()
```
